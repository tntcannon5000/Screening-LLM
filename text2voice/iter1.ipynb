```python
import torch
from TTS.api import TTS
from IPython.display import Audio, display
```

```python
# Step 4: Prepare the text you want to convert to speech
text = "According to all known laws of aviation, there is no way a bee should be able to fly. Its wings are too small to get its fat little body off the ground. The bee, of course, flies anyway because bees don't care what humans think is impossible."
```

```python
tts = TTS(model_name="tts_models/en/ljspeech/fast_pitch", progress_bar=True, gpu=True)
```

```python
output_path = "output.wav"
tts.tts_to_file(text=text, file_path=output_path)
audio_data = tts.tts(text)
```

```python
#audio_data = open(output_path, "rb").read()
#display(Audio(data=audio_data, autoplay=True))
display(Audio(audio_data, rate=22050))
```
